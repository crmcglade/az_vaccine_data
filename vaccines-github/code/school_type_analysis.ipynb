```python
import pandas as pd
pd.set_option('display.max_columns', None)

df = pd.read_csv('/Users/cmcglade/Documents/vaccines/vaccines-github/output_files/clean_kindergarten.csv')
```

## Kindergarten

```python
#check for the schools with the lowest vax rates in 2023

df23 = df[df['year'] == 2023.0]
df23.sort_values('mmr', ascending=True).head()
```

```python
#school type analysis, rates for all kids

measles_vax_school_type = df.groupby(['year', 'type'])['mmr_vax_count'].sum().to_frame(name = 'measles_vax_kids').reset_index()

totalkids_type = df.groupby(['year', 'type'])['enrolled'].sum().to_frame(name = 'total_kids').reset_index()

school_type_group = pd.merge(measles_vax_school_type, totalkids_type)
school_type_group['percent_vax'] = school_type_group['measles_vax_kids']/school_type_group['total_kids']
school_type_group.to_csv('/Users/cmcglade/Documents/vaccines/vaccines-github/output_files/kindergarten_school_type_mmr.csv')
```

```python
#school type analysis, herd immunity count

measles_vax_herd_immune = df.groupby(['year', 'type', 'mmr_herd_immune']).size().to_frame(name = 'schools').reset_index()

totalschools_type = df.groupby(['year', 'type']).size().to_frame(name = 'total_schools').reset_index()

school_type_herd = pd.merge(measles_vax_herd_immune, totalschools_type)
school_type_herd['percent_herd'] = school_type_herd['schools']/school_type_herd['total_schools']
school_type_herd = school_type_herd[school_type_herd['mmr_herd_immune'] == 'yes']

school_type_herd = school_type_herd[['year', 'type', 'schools', 'total_schools', 'percent_herd']]
school_type_herd.columns = ['year', 'type', 'herd_immune_schools', 'total_schools', 'percent_herd']
school_type_herd.to_csv('/Users/cmcglade/Documents/vaccines/vaccines-github/output_files/kindergarten_school_type_herd_mmr.csv')
```
